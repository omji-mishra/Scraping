```python
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
page_no=1
myurl = 'https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=' + str(page_no)
while(True):
    page_no = page_no + 1
    uClient = uReq(myurl)
    page_html = uClient.read()
    uClient.close()
    page_soup = soup(page_html,"html.parser")
    containers = page_soup.find_all("div",{"class":"_3O0U0u"})
    for container in containers:
        temp = container.find_all("div",{"class":"_3wU53n"})
        price = container.find_all("div",{"class":"_1vC4OE _2rQ-NK"})
        print("Price: " + price[0].text + " " + " Name: " + temp[0].text)
```
